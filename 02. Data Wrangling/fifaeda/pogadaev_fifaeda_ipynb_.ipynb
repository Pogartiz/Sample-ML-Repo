```python
ir_best_avg=df.groupby('Club').apply(lambda x:np.mean(x['International Reputation'])).reset_index(name='Average Reputation')
ir_best_cnt=df.groupby('Club').apply(lambda x:x['Name'].count()).reset_index(name='Player Counts')
ir_best_avg_cnt=pd.merge(ir_best_avg,ir_best_cnt,how='inner',left_on='Club',right_on='Club')
ir20_best_avg_cnt=ir_best_avg_cnt[ir_best_avg_cnt['Player Counts']>=20]
ir20_best_avg_cnt.sort_values(by=['Average Reputation','Player Counts'],ascending=[False,False])
fig = px.scatter(ir20_best_avg_cnt,x='Average Reputation',y='Player Counts',color='Player Counts',size='Average Reputation',hover_data=['Club'],title='Clubwise player counts and Average Reputation')
fig.show(renderer="colab")
```

Да, репутация связана с клубом, можно выделить несколько клубов, где средний рейтинг игроков ощутимо больше чем у всех остальных

Переведем столбцы Contract Valid Until и Joined в числа. Для упрощения работы будем отбрасмывать месяцы и дни, и работать только с годами.

```python
def tonumber(x):
  listx = list(x.split(' '))
  return int(listx[-1])
```

```python
df_new = df[df['Joined'].notna()]
```

```python
df_new = df_new[df_new['Contract Valid Until'].notna()]
```

```python
df_new['Joined int'] = df_new['Joined'].apply(tonumber)
```

```python
df_new['Contract Valid Until int'] = df_new['Contract Valid Until'].apply(tonumber)
```

```python
df_new[['Joined int', 'Contract Valid Until']]
```

```python
df_new['Duration'] = df_new['Contract Valid Until int'] - df_new['Joined int']
```

```python
df_new['MultV']=np.where(df_new.Value.str[-1]=='K',1000,np.where(df_new.Value.str[-1]=='M',1000000,1))
```

```python
df_new['Value']=df_new.Value.str.strip('€')
df_new['Value']=df_new.Value.str.strip('K')
df_new['Value']=df_new.Value.str.strip('M')
df_new['Value']=pd.to_numeric(df_new['Value'])
df_new['Value in €']=df_new['Value']*df_new['MultV']
```

```python
df_new_ = df_new.groupby('Duration').apply(lambda x: x['Value in €'].median()).reset_index(name='Median Value')
```

```python
fig=px.scatter(df_new_, x='Duration',y='Median Value',title='Dependency Value on Duration contract')
fig.show(renderer='colab')
```

Да, зависимость есть - цена повышается к 0-15 годам контракта, а потом падает
